```python
# importing the required module 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
data = np.loadtxt("Data/1in_tanh.txt")
imput = data[:, :1] 
output = data[:, 1:]
yHat = np.random.randn(100,1)
```

```python
# x axis values 
x = imput
# corresponding y axis values 
y1 = output
y2 = yHat
  
# plotting the points  
plt.plot(x, y1, 'blue', y2, 'red') 
  
# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
  
# giving a title to my graph 
plt.title('My first graph!') 
  
# function to show the plot 
plt.show()

#'r--', t, t**2, 'bs', t, t**3, 'g^
```
